```python
import os
import cv2
import json
import matplotlib.pyplot as plt
```

```python

# ---------- BDD100K Visualization ----------
def visualize_bdd100k_images(bdd_images_dir, bdd_labels_dir, save_dir, max_images=5):
    os.makedirs(save_dir, exist_ok=True)
    label_files = os.listdir(bdd_labels_dir)[:max_images]

    for label_file in label_files:
        with open(os.path.join(bdd_labels_dir, label_file), 'r') as f:
            data = json.load(f)
        image_path = os.path.join(bdd_images_dir, data["name"])
        img = cv2.imread(image_path)
        if img is None:
            continue

        # Draw drivable areas + boxes
        for label in data["labels"]:
            if "box2d" in label:
                box = label["box2d"]
                x1, y1, x2, y2 = map(int, [box["x1"], box["y1"], box["x2"], box["y2"]])
                cv2.rectangle(img, (x1, y1), (x2, y2), (0,255,0), 2)
                cv2.putText(img, label["category"], (x1, y1-5),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0), 1)

        out_path = os.path.join(save_dir, f"bdd_{label_file[:-5]}.jpg")
        cv2.imwrite(out_path, img)
        print(f"✅ Saved: {out_path}")

# ---------- KITTI Visualization ----------
```

```python
def visualize_kitti_labels(kitti_img_root, kitti_label_root, save_dir, max_seq=2, max_frames=10):
    os.makedirs(save_dir, exist_ok=True)
    label_files = sorted(os.listdir(kitti_label_root))[:max_seq]

    for label_file in label_files:
        seq_id = label_file.split('.')[0]
        seq_img_dir = os.path.join(kitti_img_root, seq_id)
        label_path = os.path.join(kitti_label_root, label_file)

        if not os.path.exists(seq_img_dir):
            print(f"⚠️ No images for {seq_id}")
            continue

        with open(label_path, 'r') as f:
            lines = f.readlines()

        frame_dict = {}
        for line in lines:
            parts = line.strip().split(' ')
            frame = int(parts[0])
            cls = parts[1]
            bbox = list(map(float, parts[6:10]))  # left, top, right, bottom
            frame_dict.setdefault(frame, []).append((cls, bbox))

        img_files = sorted(os.listdir(seq_img_dir))[:max_frames]
        for idx, img_name in enumerate(img_files):
            img_path = os.path.join(seq_img_dir, img_name)
            img = cv2.imread(img_path)
            if img is None:
                continue

            if idx in frame_dict:
                for cls, (x1, y1, x2, y2) in frame_dict[idx]:
                    cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0,255,0), 2)
                    cv2.putText(img, cls, (int(x1), int(y1)-5),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0), 1)

            out_path = os.path.join(save_dir, f"kitti_{seq_id}_{idx:04d}.jpg")
            cv2.imwrite(out_path, img)
            print(f"✅ Saved: {out_path}")

# ---------- MAIN ----------
```
