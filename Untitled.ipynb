```python
import os
import json
import numpy as np
import cv2
from tqdm import tqdm

label_dir = r"C:\Users\patel\Downloads\Lane_detection\bdd100k\labels\100k\train"
output_dir = r"Lane_detection\processed_lane_masks\train"
os.makedirs(output_dir, exist_ok=True)

for file in tqdm(os.listdir(label_dir)):
    if not file.endswith(".json"):
        continue
    with open(os.path.join(label_dir, file), "r") as f:
        data = json.load(f)

    mask = np.zeros((720, 1280), dtype=np.uint8)

    for obj in data.get("labels", []):
        if obj.get("category") == "lane":
            poly2d = obj.get("poly2d", [])
            for p in poly2d:
                points = np.array(p["vertices"], np.int32)
                cv2.polylines(mask, [points], isClosed=False, color=255, thickness=5)

    out_path = os.path.join(output_dir, file.replace(".json", ".png"))
    cv2.imwrite(out_path, mask)
```
